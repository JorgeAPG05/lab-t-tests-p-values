```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_rel
from scipy.stats import f_oneway
```

```python
file_path = 'machine.txt'
encoding = 'cp1252'
data_machine = pd.read_csv(file_path, header=None, names=['Current Machine', 'New Machine'], encoding=encoding)
print(data_machine.head())
```

```python
replacement_values = [
    [42.1, 42.7],
    [41, 43.6],
    [41.3, 43.8],
    [41.8, 43.3],
    [42.4, 42.5],
    [42.8, 43.5],
    [43.2, 43.1],
    [42.3, 41.7],
    [41.8, 44],
    [42.7, 44.1]
]


replacement_df = pd.DataFrame(replacement_values, columns=['Current Machine', 'New Machine'])
```

```python
data_machine.fillna(replacement_df, inplace=True)
display(data_machine)
```

```python
data_machine['Current Machine'].iloc[0] = 42.1
display(data_machine)
```

```python
data_machine.drop(index=range(10, 23), inplace=True)
data_machine.reset_index(drop=True, inplace=True)
display(data_machine)
```

```python
t_stat, p_value = ttest_rel(data_machine['New Machine'], data_machine['Current Machine'])

display("\nT-Test Results:")
display(f"T-statistic: {t_stat}")
display(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis. There is sufficient evidence to show that the new machine packs faster.")
else:
    print("\nFail to reject the null hypothesis. There is insufficient evidence to conclude that the new machine packs faster.")
```

```python
data_poke = pd.read_csv('pokemon.csv')
data_poke.head()
```

```python
column_names = data_poke.columns.tolist()

display(column_names)
```

```python
columns_to_drop= ['Sp. Atk',  'Sp. Def', 'Speed', 'Generation', 'Legendary']
data_poke1 = data_poke.drop(columns=columns_to_drop, axis=1)
display(data_poke1)
```

```python
sample = data_poke1.sample(30, random_state = 1)
sample['Difference'] = sample['Attack']-sample['Defense']
sample.head()
```

```python
t_stat, p_value = ttest_rel(sample['Attack'], sample['Defense'])

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

```python
tc = st.t.ppf(1-(0.05/2),df= sample.shape[0] - 1)
tc
```

```python
#As out T-statistic < critical value, we fail to reject H0, which means that there is no significant difference between attack and defense scores
```

Inferential statistics - ANOVA

Null hypothesis: Different plasma bean intensities do not affect the etching
Alternate hypothesis: Different plasma bean intensities affect the etching
Significance Level: 0,05
Degrees of freedom: equal to the number of groups (levels of power) minus 1. So in this case, 3
Error: equal to the total number of observations minus the number of groups. So in this case, 8
Total: Degrees of freedom equal to the total number of observations minus 11.

```python
data_anova = pd.read_excel('anova_lab_data.xlsx')
display(data_anova)
```

```python
data_anova['Power '] = data_anova['Power '].str.replace('W', '').astype(float)


display(data_anova.head())
```

```python
display(data_anova.columns)
```

```python
result_anova = f_oneway(data_anova['Power '], data_anova['Etching Rate'])

f_statistic_anova = result_anova.statistic
p_value_anova = result_anova.pvalue
print(f"\nANOVA Results:")
print(f"F-statistic: {f_statistic_anova}")
print(f"P-value: {p_value_anova}")
alpha = 0.05
if p_value_anova < alpha:
    print("\nReject the null hypothesis. There is a significant difference in the mean etching rate for different power levels.")
else:
    print("\nFail to reject the null hypothesis. There is no significant difference in the mean etching rate for different power levels.")
```
